### Question 4

```python
w = 2 * np.pi
f = 1000
w_0 = 0.25 * w
b = 0.2 * w_0
x_0 = 2
v_0 = 0
```

```python
A = f/(m.sqrt((w_0**2-w**2)**2 + 4*(b**2)*(w**2)))
delt = np.arctan((2*b*w)/((w_0**2)-(w**2)))
w_1 = m.sqrt(w_0**2 - b**2)
B1 = 2 - A*np.cos(delt)
B2 = (b*B1 - 2*np.pi*A*np.sin(delt))/w_1
```

```python
t = np.linspace(0,9,1000)
xf = A*np.cos(w*t-delt)+(m.e**(-b*t))*(B1*np.cos(w_1*t)+B2*np.sin(w_1*t))
xss = A*np.cos(w*t-delt)
xt = (m.e**(-b*t))*(B1*np.cos(w_1*t)+B2*np.sin(w_1*t))
```

```python
plt.plot(t, xf, label='Full')
plt.plot(t, xss, label='Steady-State')
plt.plot(t, xt, label='Transient')
plt.grid(alpha=0.5, linestyle='--')
plt.axhline(0, alpha=0.25, color='black')
plt.axvline(0, alpha=0.25, color='black')
plt.xlabel("Time (t)")
plt.ylabel("Amplitude (m)")
plt.title("Q4.b x(t) vs t")
plt.legend()
plt.show()
```
